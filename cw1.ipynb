#### Задание 1. Проверка СНИЛСа (0.25 балла)
Даны три строки, содержащие номера СНИЛС. Правильно записанный СНИЛС содержит 11 символов и ровно два дефиса (после первой и второй групп цифр). Проверьте три СНИЛСа на правильность и для каждого выведите True/False в зависимости от того правильная запись или нет.

```python
snils1 = '444-555-344-09' 
snils2 = '134-456-872 77' # должно вывести True
snils3 = '134 456-872-77'


def check_snils(snils):
    count_digit = sum(1 for char in snils if char.isdigit())
    return snils[3] == '-' and snils[7] == '-' and snils[11]  == ' ' and count_digit == 11

print(check_snils(snils1))
print(check_snils(snils2))
print(check_snils(snils3))
```

#### Задание 6. Угадайка (1.5 балл)

Ваша задача - воплотить игру "Угадай число". 

Программа должна генерировать случайное число от 1 до 10 (этот код мы уже написали, просто запустите его). Число генерируется в формате int.

На ввод программа принимает догадки пользователя в интервале от 1 до 10 и сравниваниет догадку со значением переменной number, которая генерируется в ячейке, которую вы запустите и выводит результат сравнения ("Вы угадали!" или "Попробуйте еще раз!). Мы предполагаем, что пользователь всегда вводит число в нужном интервале.

Программа должна работать до тех пор, пока число не будет угадано, или пока не закончатся попытки. Когда пользователь угадывает число, программа сообщает об этом и перестает работать.  

Всего у пользователя 5 попыток угадать число, после 5 неправильных попыток программа сообщает об этом и прекращает работать.

*Реализация для бесконечного числа попыток - 0.75 балла*  
*Реализация для 3 попыток - 1.5 балл*



```python
## запустите это ячейку, чтобы сгенерировать случаное число
from random import randint  # импортируем модуль random и используем функцию randint(), 
number = randint(1, 10) # которая генерирует случайное число от 1 до 10
print(number) # проверка того, какое число загадалось, чтобы вам было удобнее отлаживать программу
count = 1
while count < 4:
    resp = int(input(f'Угадайте число от 1 до 10 (осталось попыток {4 - count}): '))
    count += 1
    if resp == number:
        print(f"Урррра!! Вы угадали с {count-1} попытки!")
        break
    elif count < 4:        
        print('Попробуйте еще раз!')
    else:
        print('Неверно! Все попытки закончились. Пополните счет! :-)')
```

#### Задание 8. Магнолия (2 балла)

В магазине "Магнолия" проходит акция - даются скидки на товары, название которых начинается на заданную букву.

У Маши есть список покупок, состоящий из строк вида "цена, название, количество". Помогите ей посчитать, сколько денег она сэкономит (общую сумму скидки).

Количество записей в Машином списке покупок может быть любым. Каждая запись вводится с новой строки. Записи вводятся до тех пор, пока пользователь не введет "Конец списка". После этого одной строкой вводится буква, по которой сегодня определяется скидка. Если название продукта начинается на заданную букву, то на этот продукт действует скидка 15 процентов.

Образец вывода смотрите в примере. Если ни на один товар скидка не действует, то программа должна вывести "Скидка: 0 рублей"

*Вы считали категории в список списков, где каждая строка выглядит как в образце  - 1 балл*  
`[цена (int), название (str), количество (int)]`  
Например, `[120, мандарины (1 кг), 2]`

*Подсчет скидки - 1 балл.*

**Пример ввода**  
120 руб., мандарины (1 кг), 2 кг.  
70 руб., молоко, 3 пакета  
15 руб., батон сдобный, 1 штука  
345 руб., белое вино, 1 бутылка  
120 руб., моцарелла, 2 упаковки  
Конец списка  
м  

**Вывод**

Cкидка: 103.5 руб.


```python
def parse(line:str) -> list:
    '''Парсит входную строку в список значений.'''
    try:
        line = line.split(',')
        price = int(line[0].strip().replace('руб.', ''))
        name = line[1].strip()
        count = int(line[2].split()[0])
        return [price, name, count]
    except ValueError:
        print('Вы ввели недопустимую строку!')
    
discount = 0.15
list_purchases = []
while True:
    line_purch = input('Введите покупку: ')
    if line_purch != 'Конец списка':
        current_line = parse(line_purch)
        if current_line:
            list_purchases.append(current_line)
    else: break        
if len(list_purchases) != 0:
    char = input('Введите букву для подсчета скидки: ')
    total = sum(line[0] * line[2] * discount for line in list_purchases if line[1].startswith(char))
    print(f'Скидка: {total} руб.')
else: print('Нечего считать, пустой список покупок!')
```
